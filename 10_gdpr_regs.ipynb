```python
unique_article_numbers = df["article_number"].unique()
unique_section_numbers = df["section_number"].unique()
unique_chapter_numbers = df["chapter_number"].unique()
len(unique_article_numbers)
```

```python
import tiktoken
max_tokens = 0
model = "gpt-3.5-turbo-0613"
values = []
counter = 1
for article in unique_article_numbers:
    if article != counter:
        print(f"Article {article} should be number {counter}")
        break
    regs = get_regulation(df = df, index = str(article), index_checker = index_checker, include_section_and_chapter=False)
    encoding = tiktoken.encoding_for_model(model)
    num_tokens = len(encoding.encode(regs))
    values.append(num_tokens)
    #print(f"{article}: {num_tokens}")
    max_tokens = max(max_tokens, num_tokens)
    counter += 1
print(f"Max: {max_tokens}")
    
```

```python
indexed_values = list(enumerate(values)) 
indexed_values.sort(key=lambda x:x[1])
```

```python
g = 800
count_exceeds_g = sum(value > g for index, value in indexed_values)
count_exceeds_g
```
